## Tính toán tương quan cho từng kỳ và cả năm

### Subtask:
Lặp qua từng môn học và tính toán hệ số tương quan giữa điểm của từng kỳ (s1, s2) và điểm cả năm (y) với điểm HSA.


**Reasoning**:
I need to iterate through each grade level and then each subject within that grade, calculate the correlation between each semester and year score with the HSA score, and store these correlations in a dictionary.



```python
# Initialize an empty dictionary to store correlation results
correlation_results = {}

# Define the list of grades
grades = ['g10', 'g11', 'g12']

# Iterate through each grade
for grade in grades:
    # Get the list of subject columns for the current grade
    subject_columns = subject_gpa_cols[grade]

    # Iterate through the subject columns in steps of 3 (s1, s2, y for each subject)
    for i in range(0, len(subject_columns), 3):
        # Extract the subject name (e.g., 'math' from 'g10_math_s1')
        # Assuming the format is always grade_subject_semester/year
        subject_name = subject_columns[i].split('_')[1]

        # Define the column names for semester 1, semester 2, and year scores
        s1_col = subject_columns[i]
        s2_col = subject_columns[i+1]
        y_col = subject_columns[i+2]

        # Calculate the correlation with hsa_score for each period
        if s1_col in df.columns and hsa_score_col in df.columns:
            correlation_s1 = df[s1_col].corr(df[hsa_score_col])
            correlation_results[f'{grade}_{subject_name}_s1'] = correlation_s1

        if s2_col in df.columns and hsa_score_col in df.columns:
            correlation_s2 = df[s2_col].corr(df[hsa_score_col])
            correlation_results[f'{grade}_{subject_name}_s2'] = correlation_s2

        if y_col in df.columns and hsa_score_col in df.columns:
            correlation_y = df[y_col].corr(df[hsa_score_col])
            correlation_results[f'{grade}_{subject_name}_y'] = correlation_y

# Display the dictionary containing all calculated correlation coefficients
display(correlation_results)
```

## Trực quan hóa kết quả tương quan

### Subtask:
Tạo biểu đồ (ví dụ: heatmap) để hiển thị ma trận tương quan giữa điểm các môn theo từng kỳ, cả năm và điểm HSA, sử dụng màu xanh lá cây và đỏ.


**Reasoning**:
Convert the correlation results dictionary to a DataFrame and sort it, then create a heatmap to visualize the correlations using a diverging color map as requested by the subtask.



```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def corr_(da):
    df_corr = da.corr()
    ones_corr = np.ones_like(df_corr, dtype=bool)
    # np's triu: return only upper triangle matrix
    mask = np.triu(ones_corr)
    adjusted_mask = mask[1:, :-1]
    adjusted_dsts_corr = df_corr.iloc[1:, :-1]

    # Increase the figure size for a larger heatmap
    plt.figure(figsize=(10, 8)) # Increased figure size

    cmap = sns.diverging_palette(0, 125, 100, 50, as_cmap=True)
    sns.heatmap(data=adjusted_dsts_corr, mask=adjusted_mask, cmap=cmap,
                vmin=0, vmax=1, linecolor="white", linewidths=0.5, annot=True, fmt=".2f")
    plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
    plt.yticks(rotation=0) # Ensure y-axis labels are horizontal
    plt.tight_layout() # Adjust layout to prevent labels overlapping
    plt.show()
```

## Huấn luyện và đánh giá mô hình mạng nơ-ron nhân tạo (ann)

### Subtask:
Xây dựng, huấn luyện mô hình ANN trên tập huấn luyện, dự đoán trên tập kiểm tra và tính toán các chỉ số đánh giá (MSE, RMSE, MAPE).


**Reasoning**:
Import necessary libraries for building and training the ANN model, build the sequential ANN model, compile it with an optimizer and loss function, and train the model on the scaled training data. Then, make predictions on the scaled test data and calculate the evaluation metrics (MSE, RMSE, MAPE).



```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
import numpy as np

# Define the number of features
n_features = X_train_scaled.shape[1]

# Build the ANN model
ann_model = Sequential()
ann_model.add(Dense(128, activation='relu', input_shape=(n_features,)))
ann_model.add(Dropout(0.2)) # Added Dropout layer
ann_model.add(Dense(64, activation='relu'))
ann_model.add(Dropout(0.2)) # Added Dropout layer
ann_model.add(Dense(32, activation='relu'))
ann_model.add(Dense(1)) # Output layer for regression

# Compile the model
ann_model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mse', 'mae'])

# Train the model
history = ann_model.fit(X_train_scaled, y_train, epochs=50, batch_size=32, validation_split=0.2, verbose=0) # Added validation_split and set verbose to 0

# Make predictions on the scaled test set
y_pred_ann = ann_model.predict(X_test_scaled).flatten() # Flatten predictions to match y_test shape

# Calculate evaluation metrics
mse_ann = mean_squared_error(y_test, y_pred_ann)
rmse_ann = np.sqrt(mse_ann)
mape_ann = mean_absolute_percentage_error(y_test, y_pred_ann)

# Print the calculated metrics
print(f"ANN Model - MSE: {mse_ann:.4f}")
print(f"ANN Model - RMSE: {rmse_ann:.4f}")
print(f"ANN Model - MAPE: {mape_ann:.4f}")
```
